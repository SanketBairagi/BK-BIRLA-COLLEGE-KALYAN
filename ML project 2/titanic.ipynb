```python
df = pd.read_csv("https://raw.githubusercontent.com/ingledarshan/BK_Birla/main/titanic.csv")
df.shape
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.drop(['Cabin'], axis=1, inplace=True)
df.isna().sum()
```

```python
df.Embarked.fillna(df.Embarked.mode()[0], inplace=True)
df.isna().sum().sort_values(ascending=False)
```

```python
df.Age.fillna(df.Age.median(), inplace=True)
df.isna().sum().sort_values(ascending=False)
```

```python
df['FamilySize']  = df['SibSp'] + df['Parch'] + 1
df.head()
```

```python
df['GenderClass'] = df.apply(lambda x : 'child' if x['Age']<15 else x['Sex'], axis=1)
df.sample(10)
```

```python
df.drop(['PassengerId', 'Ticket','Name',"Sex"], axis=1, inplace=True)
df.head(15)
```

```python
df.drop(['SibSp', 'Parch'], axis=1, inplace=True)
df.head()
```

```python
GenderClass = {'female':0, 'male':1,'child':2}
df['GenderClass'] = df['GenderClass'].replace(GenderClass)
```

```python
df.head(10)
```

```python
Embarked = {'C':0, 'Q':1,'S':2}
df['Embarked'] = df['Embarked'].replace(Embarked)
```

```python
df.head(10)
```

```python
X = df.loc[:,df.columns != 'Survived']
X.head()
```

```python
y = df.Survived
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred_train = logreg.predict(X_train)
y_pred_test = logreg.predict(X_test)
```

```python
pd.DataFrame(zip(y_train, y_pred_train), columns=['Actual','Predicted'])
```

```python
pd.DataFrame(zip(y_test, y_pred_test), columns=['Actual','Predicted'])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Accuracy on train data
accuracy_score(y_train, y_pred_train)
```

```python
# Accuracy on test data
accuracy_score(y_test, y_pred_test)
```

```python
confusion_matrix(y_test, y_pred_test)
```

```python
cm = pd.DataFrame(confusion_matrix(y_test, y_pred_test))
cm
```

```python
cm.index = ['Actual Died', 'Actual Survived']
cm.columns = ['Predicted Died', 'Predicted Survived']
cm
```
