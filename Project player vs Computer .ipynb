4- in this code first it takes the name of player also number of rounds from the user, then all the coditions getting checked. as the condition statisfy, its assigned the score to player and computer.comparing nomber of score the winner is desided.    


```python
### NAME OF COMPUTER IN GAME IS "Jarvis"
import random
g = 2
while g > 1:
    print("COLOR CODE :- 1 - RED \t 2 - YELLOW \t 3 - ORENGE \t 4 - GREEN \t 5 - BLUE")
    name = input("\nENTER YOUR NAME - ")
    while True:
        try:
            x = int(input("ENTER NUMBER OF ROUNDS : "))
            if x >=1:
                break;
            else:
                print("INVALID NUMBER , TRY AGAIN (NUMBER OF ROUND SHOULD BE GREATER THAN 0 )")
        except ValueError:
            continue
    n = 1
    countp = 0
    countc = 0
    while n <= x:
        print("\nROUND-", n)
        while True:
            try:
                a = int(input("ENTER COLOR CODE : "))
                if a >= 1 and a <= 5:
                    break;
                else:
                    print("INVALID NUMBER , TRY AGAIN ")
            except ValueError:
                continue

        b = random.randint(1, 5)
        print(name,"-",a)
        print("Jarvis -", b)

        if a == b:
            countp = countp + 1
        if a < b or a > b:
            countc = countc + 1

        print("\nROUND ",n," SCORE : ")
        print(name,":", countp,end="     ")
        print("Jarvis :", countc)
        print("\n********************************************")
        n = n + 1
    if countp > countc:
        print("\n\n---- FINAL RESULT ----")
        print(name, " WIN")
        print("Jarvis LOST\n\n")
    elif countp < countc:
        print("\n\n---- FINAL RESULT ----")
        print("Jarvis WIN")
        print(name," LOST\n\n")
    else:
        print("\n\nGAME TYED\n\n")

    while True:
        try:
            h = int(input("you want to play again?\n Enter 1 - YES\n ENTER 2 - NO \n ------> "))
            if h == 1 :
                print("==========================================================================================")
                break;
            if h==2:
                print("THANK YOU FOR PLAYING GAME")
                g=1
                break;
            else:
                print("invalid number , try again")
        except ValueError:
            continue

```
